```python
#Missoin_2.  20개의 정수를 입력 받아서 짝수와 홀수를 구분하여 저장해주세요. 
#           저장된 수의 총합이 얼마인지와 짝수의 합이 큰지 홀수의 합이 큰지를 알려주는 코드를 작성해주세요 

even_numbers = []
#짝수 리스트 생성
odd_numbers = []
#홀수 리스트 생성

for i in range(20):
#변수 i범위 지정(0-20까지 반복)
    num = int(input("정수 입력하기(20개) : "))
    #num 값 입력받음
    if num % 2 == 0:
    #만약 입력된 수가 짝수라면
        even_numbers.append(num)
        #짝수 리스트에 입력된 수를 추가한다
    else:
    #짝수 아니면=홀수라면,
        odd_numbers.append(num)
        #홀수 리스트에 입력된 수를 추가한다

even_sum = sum(even_numbers)
#짝수 총합
odd_sum = sum(odd_numbers)
#홀수 총합

print("짝수", even_numbers)
#입력된 짝수를 모두 출력
print("홀수", odd_numbers)
#입력된 홀수를 모두 출력
print("짝수의 총합 : ", even_sum)
print("홀수의 총합 : ", odd_sum)
print("모든 수의 총합 : ", even_sum + odd_sum)
#짝수 총합, 홀수 총합, 입력된 모든 수의 총합 출력

if even_sum > odd_sum:
#만약 짝수의 총합이 크다면,
    print("짝수의 합이 더 큼")
    #"짝수 합이 더 큼"을 출력
elif odd_sum > even_sum:
#만약 홀수의 합이 더 크다면,
    print("홀수의 합이 더 큼")
    #홀수 합이 더 큼을 출력
else:
#아니라면(어느 쪽도 더 크지 않다면=두 합이 같다면)
    print("짝수와 홀수의 합이 같음")
    #"짝수 홀수의 합이 같음"을 출력
    
#도저히 혼자서 못하겠어서 챗 GPT 썼습니다.
```

```python
#Mission_9. 가위,바위,보 게임을 만들어주세요.. (이긴 횟수를 알려주세요.)

import random
#랜덤 모듈 가져오기

def rps_game(user, computer):
#가위바위보 게임 함수를 정의
    if user == computer:
    #만약 사용자와 컴퓨터가 같은 것을 내면
        return "무승부"
        #"무승부"라는 문자열 반환
    elif (
        (user == "가위" and computer == "보") or
        (user == "바위" and computer == "가위") or
        (user == "보" and computer == "바위")
    ):
    #만약 사용자가 가위를 내고 컴퓨터가 보를 내거나
    #또는, 사용자가 바위를 내고 컴퓨터가 가위를 내거나
    #또는, 사용자가 보를 내고 컴퓨터가 바위를 내면
        return "인간 승리"
        #"인간 승리"라는 문자열 반환
    else:
    #아니면,
        return "컴퓨터 승리"
        #"컴퓨터 승리"라는 문자열 반환

user_wins = 0
#사용자가 이긴 횟수
total_games = 0
#총 게임 횟수

while True:
    choices = ["가위", "바위", "보"]
    #리스트에 가위, 바위, 보 저장
    user = input("가위, 바위, 보 중에서 하나를 선택하세요(종료하려면 q입력) : ")
    #사용자로부터 가위, 바위, 보 중 하나를 입력 받음
    
    if user == "q":
    #만약 사용자가 q입력하면
        print("게임 종료")
        #게임 종료 반환하고
        break
        #무한 루프 빠져나감
    
    if user not in choices:
    #만약 사용자가 choices 리스트에 없는 요소 입력
        print("잘못 입력하셨습니다. 다시 입력하세요.")
        #다시 입력하라는 메시지 반환하고
        continue
        #계속 진행
    
    computer = random.choice(choices)
    #컴퓨터는 choices 리스트에서 무작위로 선택함
    print("인간의 선택 :", user)
    #사용자 입력을 반환
    print("컴퓨터의 선택 :", computer)
    #컴퓨터 선택을 반환
    
    winner = rps_game(user, computer)
    #rps_game 함수 호출. 둘 중 이긴 플레이어 반환해 winner 변수에 저장한다.
    print("결과 :", winner)
    #이긴 사람 반환
    print()
    
    if winner == "인간 승리":
    #만약 사용자가 승리하면
        user_wins += 1
        #사용자 승리 변수에 1 누적
    total_games += 1
    #총 게임 횟수 변수에 1 누적
    
print("이긴 횟수 :", user_wins)
#이긴 횟수 출력
print("총 게임 횟수 :", total_games)
#총 게임 횟수 출력
```

```python
#Mission_10. 커피 자판기의 계산프로그램을 만들려고 합니다.(최대 15잔을 뽑을 수 있어요. 잔당 200원)
#            돈을 값을 입력 받으면 커피를 뽑을 수 있는 잔수와 잔액을 알려주시면 됩니다. 

def coffee_machine(money):
#커피머신 만들기 함수 정의
    
    coffee_price = 200
    #커피 가격은 200원
    max_cups = min(money // coffee_price, 15)
    #최대잔수 = 머신에 넣은 돈을 커피 가격으로 나눈 몫
    #커피는 15잔까지만 뽑을 수 있음 즉, 몫은 15 이하여야 함
    change = money % coffee_price
    #거스름돈 = 머신에 넣은 돈을 커피 가격으로 나눈 값의 나머지

    if money >= coffee_price:
    #만약 가진 돈이 커피 가격 이상이라면,
        return max_cups, change
        #최대 커피 잔 수와, 거스름돈 값을 반환함
    else:
        return "돈이 부족하여 커피를 마실 수가 없습니다."
        #200원 미만을 입력하면 반환하는 메시지
        
money = int(input("얼마나 가졌느냐??? : "))
result = coffee_machine(money)
 
if len(result) == 2:
#만약 변수 result 길이가 2이면,
    max_cups, change = result
    #최대 잔수와 거스름돈 반환
    print("최대로 뽑을 있는 커피 잔 수: ", max_cups)
    print("거스름돈 : ", change, "원")
else:
    print(result)

#챗GPT가 만들어 줬어요
```
